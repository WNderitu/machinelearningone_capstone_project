## 3.3 Data visualization

### 3.3.1 Creating functions for visualization

#### Plot stacked bars & annotate stacked bar functions

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):   
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectangles/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

#### Distribution plot function

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distribution in a stacked histogram of churned or retained customer
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

#### boxplot function

```python
def plot_boxplot(dataframe, column, ax):
    """
    Plot a box plot for a given column in the dataframe with churn labels.
    
    Parameters:
    dataframe (pd.DataFrame): The dataframe containing the data.
    column (str): The column name for which the box plot is to be plotted.
    ax (matplotlib.axes._axes.Axes): The axes on which to plot the box plot.
    """
    
    sns.boxplot(data=dataframe, x='churn', hue = 'churn',  y=column, ax=ax, legend= False)
    ax.set_title(f'Box plot of {column} by churn')
    ax.set_xlabel('Churn')
    ax.set_ylabel(column)
    ax.set_xticks([0, 1])  # Set the ticks explicitly
    ax.set_xticklabels(['Retained', 'Churn'])  # Set the tick labels
```

#### KDE plot function

```python
def plot_kde(dataframe, column, ax):
    """
    Plot a Kernel Density Estimate (KDE) plot for a given column in the dataframe.
    
    Parameters:
    dataframe (pd.DataFrame): The dataframe containing the data.
    column (str): The column name for which the KDE plot is to be plotted.
    ax (matplotlib.axes._axes.Axes): The axes on which to plot the KDE plot.
    """
    
    sns.kdeplot(data=dataframe[dataframe['churn'] == 0], x=column, fill=True, ax=ax, label='Retained')
    sns.kdeplot(data=dataframe[dataframe['churn'] == 1], x=column, fill=True, ax=ax, label='Churn')
    ax.set_title(f'KDE plot of {column} by churn')
    ax.set_xlabel(column)
    ax.set_ylabel('Density')
    ax.legend(loc='upper right')
```

### 3.3.4 Analysis of the categorical columns per churn label

```python
# use groupby function to group the channel_sales data by churn label
# create a dataframe containing the churn, id and channel_sales columns
channel_sales_df = client_df[['id','channel_sales', 'churn']]

# group the channel_sales_data by churn label
channel_sales_data = channel_sales_df.groupby(['channel_sales','churn'])['id'].count()
channel_sales_data = channel_sales_data.unstack(level=1)
channel_sales_data = channel_sales_data.fillna(0)
# calculate the percentage of churn and retention
channel_sales_data = channel_sales_data.div(channel_sales_data.sum(axis=1), axis=0) * 100 
# arrange in descending order
channel_sales_data = channel_sales_data.sort_values(by=1, ascending=False)

plot_stacked_bars(channel_sales_data, "Channel Sales", (20,5), legend_="upper right", rot_=30)
```

Insights:
1. There are 3 sales channels with no churned companies

```python
# use groupby function to group the origin_up data by churn label. 
# origin_up is the code of electricity campaign the customer first subscribed to.
# create a dataframe containing the churn, id and origin_up columns
code_df = client_df[['id','origin_up', 'churn']]

# group the origin_up data by churn label
code_data = code_df.groupby(['origin_up', 'churn'])['id'].count().unstack(level=1).fillna(0)
# calculate the percentage of churn and retention
code_data = code_data.div(code_data.sum(axis=1), axis=0) * 100
# arrange in descending order
code_data = code_data.sort_values(by=1, ascending=False)

plot_stacked_bars(code_data, "Code of electricity campaign", (20, 5), legend_="upper right", rot_=30)
```

Insights:

1. There are two codes of the electricity campaign that companies didn't churn.

```python
# use groupby function to group the has_gas data by churn label. 
# create a dataframe containing the churn, id and has_gas columns
gas_df = client_df[['id','has_gas', 'churn']]
# group the has_gas data by churn label
gas_data = gas_df.groupby(['has_gas','churn'])['id'].count().unstack(level=1).fillna(0)
# calculate the percentage of churn and retention
gas_data = gas_data.div(gas_data.sum(axis=1), axis=0) * 100
# arrange in descending order
gas_data = gas_data.sort_values(by=1, ascending=False)

plot_stacked_bars(gas_data, "Has Gas", (8, 4), legend_="upper right", rot_=0)
```

Insights:
